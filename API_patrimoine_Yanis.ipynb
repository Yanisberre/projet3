```python
import pandas as pd
import requests
from pprint import pprint
import time
from tqdm import tqdm
import plotly.express as px
```

```python
url ="https://data.iledefrance.fr/api/explore/v2.1/catalog/datasets/sites-labellises-patrimoine-dinteret-regional/records?limit=20&offset{depart}"

r = requests.get(url)
print(r)
r_js = r.json()
df = pd.json_normalize(r_js, record_path = "results")
print(df.head(100).to_markdown())
```

```python
l = []
for depart in tqdm(range(0,218,20)):
  url = f"https://data.iledefrance.fr/api/explore/v2.1/catalog/datasets/evenements-publics-cibul/records?limit=20&offset={depart}&refine=lastdate_end%3A%222024%22&refine=updatedat%3A%222024%22&refine=firstdate_begin%3A%222024%22"
  r = requests.get(url)
  r_js = r.json()
  df = pd.json_normalize(r_js, record_path = "results")
  l.append(df)
df_all_pat  = pd.concat(l)
###
```
